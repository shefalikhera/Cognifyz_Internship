```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/Dataset .csv')
```

```python
df
```

```python
# cleaning of the dataset

df.isnull().sum()

df.dropna(inplace=True)

df.duplicated().sum()

df.drop_duplicates(inplace=True)
```

```python
def remove_outliers(df, col):

    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
```

```python
df
```

```python
#Top Cuisines
cuisine_counts = df['Cuisines'].value_counts()
top_3_cuisines = cuisine_counts.head(3)

total_restaurants = len(df)
top_3_cuisine_percentages = (top_3_cuisines / total_restaurants) * 100
print("Top 3 Most Common Cuisines:")
print(top_3_cuisines)
print("\nPercentage of Restaurants Serving Each Cuisine:")
print(top_3_cuisine_percentages)
```

```python
#City Analysis
city_with_most_restaurants = df['City'].value_counts().idxmax()

average_ratings_by_city = df.groupby('City')['Aggregate rating'].mean()

city_with_highest_average_rating = average_ratings_by_city.sort_values(ascending=False).index[0]

print(f"City with the most restaurants: {city_with_most_restaurants}")
print(f"average rating for restraunts: {average_ratings_by_city}")
print(f"City with highest average rating: {city_with_highest_average_rating}")
```

```python
#PRICE RANGE DISTRIBUTION
import matplotlib.pyplot as plt
%matplotlib inline
price_range_counts = df['Price range'].value_counts()
total_restaurants = len(df)
price_range_percentages = (price_range_counts / total_restaurants) * 100

price_range_percentages.plot(kind='bar')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.title('Distribution of Price Ranges')
plt.show()
```

```python
#Online Delivery
delivery_counts = df['Has Online delivery'].value_counts()
total_restaurants = len(df)
delivery_percentage = (delivery_counts[1] / total_restaurants) * 100
print("Percentage of restaurants offering online delivery:", delivery_percentage)

delivery_ratings = df.groupby('Has Online delivery')['Aggregate rating'].mean()
print("Average rating of restaurants with online delivery:", delivery_ratings[1])
print("Average rating of restaurants without online delivery:", delivery_ratings[0])
```
