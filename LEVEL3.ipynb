```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Dataset  (2).csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
df.dropna(inplace= True)
df.duplicated().sum()
df.drop_duplicates(inplace= True)
```

```python
df
```

```python
#Votes Analysis (TASK 2)
df.groupby('Restaurant Name')['Votes'].agg(['min', 'max'])
```

```python
correlation = df['Votes'].corr(df['Aggregate rating'])
print(f"The correlation between votes and rating is {correlation:.2f}")

# Create a scatter plot to visualize the relationship
plt.scatter(df['Votes'], df['Aggregate rating'])
plt.xlabel('Votes')
plt.ylabel('Aggregate rating')
plt.title('Scatter Plot of Votes vs Rating')
plt.show()
```

```python
#Price Range vs. Online Delivery and Table Booking( TASK 3)
grouped_data = df.groupby(['Price range', 'Has Online delivery'])['Has Table booking'].count()
print(grouped_data)

grouped_data.unstack().plot(kind='bar')
plt.xlabel('Price Range and Online delivery')
plt.ylabel('Count of Restaurants')
plt.title('Relationship between Price Range, Online Delivery, and Table Booking')
plt.show()
```
